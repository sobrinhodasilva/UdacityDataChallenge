#### Análise exploratoria indadimplência

#### Modelo de risco de crédito 
- remover os target sem resultado (NaN)
- target_default true/false to int 1/0 
- score 1,2,3 NaN to 0
- n_defaulted_loans NaN to 0 


#### Code

Dados treino aquisicao para o modelo de risco de credito 

```python
#selecionar as colunas 
df_risco_train = df_ac_train[['ids', 'target_default', 'score_1', 'score_2', 'score_3', 'score_4', 'score_5' \
                            , 'score_6','risk_rate','n_defaulted_loans']]

df_risco_test = df_ac_test[['ids', 'score_1', 'score_2', 'score_3', 'score_4', 'score_5'\
                            , 'score_6','risk_rate','n_defaulted_loans']]
```

```python
#remover nan target_default
df_risco_train = df_risco_train.query('target_default.notnull()')

#substituir target_default true/false to int 1/0
df_risco_train.target_default = df_risco_train.target_default * 1
```

```python
#criar outro dataset sem os nulos 
df_risco_train_notNull = df_risco_train
df_risco_test_notNull = df_risco_test
#score 3 NaN to 0 
df_risco_train_notNull.score_3 = df_risco_train_notNull.score_3.fillna(0)

df_risco_test_notNull.score_3 = df_risco_test_notNull.score_3.fillna(0)
```

```python
#n_defaulted_loans NaN to 0 
df_risco_train_notNull.n_defaulted_loans = df_risco_train.n_defaulted_loans.fillna(0)

df_risco_test_notNull.n_defaulted_loans = df_risco_test.n_defaulted_loans.fillna(0)
```

```python
df_risco_train_notNull.risk_rate = df_risco_train_notNull.risk_rate.fillna(0)

df_risco_test_notNull['risk_rate'] = df_risco_test_notNull.risk_rate.fillna(0)
```

```python
df_risco_test_notNull['score_1'] = df_risco_test_notNull.score_1.fillna(0)
df_risco_test_notNull['score_2'] = df_risco_test_notNull.score_2.fillna(0)
```
